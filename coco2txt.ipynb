```python
import os
import random
import scipy
import shutil
from pathlib import Path
from PIL import Image, ImageDraw
import json
```

```python
with open("custom_test1525_prediction(coco_format)(sh=0.3)(1).json", "r") as open_f:
    raw_data = json.load(open_f)
```

```python
label_dir = "datasets/cp/labels/test"
img_dir = "datasets/cp/images/test"
```

```python
for image in raw_data["images"]:
    id, file_name, img_width, img_height = (
        image["id"],
        image["file_name"],
        image["width"],
        image["height"],
    )

    label_lines = []
    for annotation in raw_data["annotations"]:
        if annotation["image_id"] == id:
            upperleft_x, upperleft_y, width, height = annotation["bbox"]
            center_x = upperleft_x + width / 2
            center_y = upperleft_y + height / 2

            center_x = str(round(center_x / img_width, 6))
            center_y = str(round(center_y / img_height, 6))
            width = str(round(width / img_width, 6))
            height = str(round(height / img_height, 6))

            label_lines.append(" ".join([str(0), center_x, center_y, width, height]))

    label_dst = os.path.join(label_dir, file_name.split(".")[0] + ".txt")
    with open(label_dst, "w") as file:
        file.write("\n".join(label_lines))
```

```python
rand_index = random.randint(0, len(raw_data["images"]))
print(rand_index)

img_name = os.path.join(img_dir, raw_data["images"][rand_index]["file_name"])
label_name = os.path.join(
    label_dir, raw_data["images"][rand_index]["file_name"].split(".")[0] + ".txt"
)
with open(label_name, "r") as file:
    labels = file.read().split("\n")

print(labels)

image = Image.open(img_name)

draw = ImageDraw.Draw(image, "RGBA")
for box in labels:
    class_id, x, y, w, h = box.split(" ")

    w = eval(w) * image.width
    h = eval(h) * image.height
    x = eval(x) * image.width - w / 2
    y = eval(y) * image.height - h / 2

    print(class_id, x, y, w, h)

    draw.rectangle((x, y, x + w, y + h), outline="red", width=4)

image
```
